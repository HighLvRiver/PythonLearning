scatterplot 그려보자!

```python
iris.plot(kind="scatter", x="SepalLengthCm", y="SepalWidthCm")
plt.show()
```

seaborn 을 사용해서도 비슷한 plot을 그릴 수 있다. seaborn의 경우, bivariate scatterplot 과 univariate scatterplot을 한 figure에 합쳐서 나타낼 수 있다.

```python
sns.jointplot(x="SepalLengthCm", y="SepalWidthCm", data=iris, size=5)
plt.show()
```

위에서 그린 scatterplot에는 종에 대한 정보가 빠져 있으므로, 색상으로 종을 구분해보고자한다

```python
#sns.FacetGrid(iris, hue="Species", size=5).map(plt.scatter, "SepalLengthCm", "SepalWidthCm").add_legend()

sns.FacetGrid(iris, hue="Species", size=5) \
   .map(plt.scatter, "SepalLengthCm", "SepalWidthCm") \
   .add_legend() #범례

plt.show()
```

각각의 종별로 box-plot으로도 볼까요?

```python
sns.boxplot(x="Species", y="PetalLengthCm", data=iris)
plt.show()
```

```python
ax = sns.boxplot(x="Species", y="PetalLengthCm", data=iris)
ax = sns.stripplot(x="Species", y="PetalLengthCm", data=iris, jitter=True, edgecolor="gray")

plt.show()
```

```python
sns.violinplot(x="Species", y="PetalLengthCm", data=iris, size=6)
plt.show()
```

Kernel Density Estimation
- http://darkpgmr.tistory.com/147

```python
sns.FacetGrid(iris, hue="Species", size=6) \
   .map(sns.kdeplot, "PetalLengthCm") \
   .add_legend()

plt.show()
```

```python
sns.pairplot(iris.drop("Id", axis=1), hue="Species", size=3)
plt.show()
```

```python
sns.pairplot(iris.drop("Id", axis=1), hue="Species", size=3, diag_kind="kde")
plt.show()
```

```python
iris.drop("Id", axis=1).boxplot(by="Species", figsize=(12, 6))
plt.show()
```

Andrews Curves
- Andrews Curves involve using attributes of samples as coefficients for Fourier series
- https://en.wikipedia.org/wiki/Andrews_plot

```python
from pandas.tools.plotting import andrews_curves
andrews_curves(iris.drop("Id", axis=1), "Species")
plt.show()
```

Parallel coordinates
- https://en.wikipedia.org/wiki/Parallel_coordinates

```python
from pandas.tools.plotting import parallel_coordinates
parallel_coordinates(iris.drop("Id", axis=1), "Species")

plt.show()
```

Radviz

```python
from pandas.tools.plotting import radviz
radviz(iris.drop("Id", axis=1), "Species")

plt.show()
```
